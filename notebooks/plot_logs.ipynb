```python
filtered_df = tottime_df[tottime_df['tottime_16'] < tottime_df['tottime_128']]
filtered_df
```

```python
def seconds_to_hours_minutes(seconds):
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    return hours, minutes
```

```python
y=tottime_df[tottime_df.fname=="<method 'read' of '_ssl._SSLSocket' objects>"][["tottime_16","tottime_32","tottime_64","tottime_128"]]
y=y.values.flatten()
hy = [seconds_to_hours_minutes(yy) for yy in y]
my = [h[1] for h in hy]
x=[16,32,64,128]
xlabels = ["16","32","64","128"]
plt.figure()
plt.xticks(x,xlabels)
plt.xlabel('max cuncurrent connections')
plt.ylabel('time (minutes)')
plt.plot(x,my,marker='*',linestyle='--',linewidth=0.7,markersize=7,label='read SSL socket')


y2=tottime_df[tottime_df.fname=="<method 'poll' of 'select.epoll' objects>"][["tottime_16","tottime_32","tottime_64","tottime_128"]]
y2=y2.values.flatten()
hy2 = [seconds_to_hours_minutes(yy) for yy in y2]
my2 = [h[1] for h in hy2]
plt.plot(x,my2,marker='*',linestyle='--',linewidth=0.7,markersize=7,label='monitor SSL socket')

y3=tottime_df[tottime_df.fname=="<method 'join' of 'bytes' objects>"][["tottime_16","tottime_32","tottime_64","tottime_128"]]
y3=y3.values.flatten()
hy3 = [seconds_to_hours_minutes(yy) for yy in y3]
my3 = [h[1] for h in hy3]
plt.plot(x,my3,marker='*',linestyle='--',linewidth=0.7,markersize=7,label='join received bytes')

y4=tottime_df[tottime_df.fname=="<method 'write' of '_ssl.MemoryBIO' objects>"][["tottime_16","tottime_32","tottime_64","tottime_128"]]
y4=y4.values.flatten()
hy4 = [seconds_to_hours_minutes(yy) for yy in y4]
my4 = [h[1] for h in hy4]
plt.plot(x,my4,marker='*',linestyle='--',linewidth=0.7,markersize=7,label='write data into memory buffer')

plt.legend()
```
