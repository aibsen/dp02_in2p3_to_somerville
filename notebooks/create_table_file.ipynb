```python
from astropy.table import Table
import pandas as pd
```

```python
urls = pd.read_csv('../csv_files/deepCoadd_calexp_urls.csv', header=None, names=['urls'])
a = urls.head()
```

```python
base_dir = '/data/butler/dp02/'
```

```python
def change_base_directory(url):
    directories=url.split('/')[3:]
    band = directories[-2]
    patch = int(directories[-3])
    tract = int(directories[-4])
    skymap = 'DC2'
    run = directories[-6]
    new_url = base_dir+'/'.join(directories)
    return new_url, band, skymap, tract, patch, run
```

```python
change_base_directory(urls.iloc[1].values[0])
```

```python
urls['filename'], urls['band'], urls['skymap'], urls['tract'], urls['patch'], urls['run']= zip(*urls['urls'].apply(lambda x: change_base_directory(x)))
```

```python
by_run = urls.groupby('run')
```

```python
for run_value, group_df in by_run:
    new_group = group_df[['filename', 'band', 'skymap', 'tract', 'patch']]
    file_table = Table.from_pandas(new_group)
    # print(file_table)
    file_table.write(f'../deepCoadds_calexp_{run_value}.ecsv')
```
